```python
import pandas as pd
```

```python
from sqlalchemy import create_engine

engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
import pyarrow.parquet as pq

parquet_file = pq.ParquetFile("./green_tripdata_2025-11.parquet")
```

```python
first = True

print("Starting db import")

for batch in parquet_file.iter_batches():
    batch_df = batch.to_pandas()
    if first:
        batch_df.head(0).to_sql(
            name="green_trip_data",
            con = engine,
            if_exists="replace"
        )
        print("db created")
    batch_df.to_sql(
            name="green_trip_data",
            con = engine,
            if_exists="append"
    )
    print(f"Inserted #{len(batch_df)} at green_trip_data")
print("done!")
    
        
```
